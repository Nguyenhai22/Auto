```python
from sqlalchemy import create_engine
import pandas as pd
import urllib.parse
import os

db_config = {
    "host": "192.168.0.222",  
    "user": "mydata",  
    "password": "mydata@2024",
    "database": "tosypro_lvl"  
}
```

```python
table_list = [
    "`tosypro_lvl`.`export_materials`", 
    "`tosypro_lvl`.`import_warehouses`", 
    # "`tosypro_lvl`.`materials`", 
    "`tosypro_lvl`.`pack_box_materials`"
    # "`tosypro_lvl`.`material_processes`"
    ]

base_output_folder = r"C:\Users\trungnt\OneDrive - tosy.com\DATA ENGINEERING - Other\NguyenHai\Data_Trộn nhựa"

def extract_table_name(full_table_name):
    return full_table_name.split(".")[1].replace("`", "")
encoded_password = urllib.parse.quote_plus(db_config['password'])
connection_string = f"mysql+pymysql://{db_config['user']}:{encoded_password}@{db_config['host']}/{db_config['database']}"

try:
    engine = create_engine(connection_string)
    print("Kết nối thành công!")

    os.makedirs(base_output_folder, exist_ok=True)

    for table in table_list:
        try:
            table_name = extract_table_name(table)
            print(f"Đang xử lý: {table_name}")
            query = f"SELECT * FROM {table} WHERE created_at >= '2026-01-01'"
            df = pd.read_sql(query, con=engine)
            file_name = f"{table_name}.csv"
            output_csv = os.path.join(base_output_folder, file_name)
            df.to_csv(output_csv, index=False, encoding="utf-8-sig")
            print(f"Đã lưu: {output_csv}")
        except Exception as e:
            print(f"Lỗi khi xử lý bảng {table}: {e}")

    print("Hoàn thành xử lý!")

except Exception as e:
    print(f"Lỗi: {e}")
```
